```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/Position Salary.csv")
df
```

```python
x=df[["Level"]]
y=df[["Salary"]]
```

Model Building

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x,y)
```

```python
y_pred=regressor.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
x_poly=PolynomialFeatures(degree=2)
```

```python
#fit method works for parameters to calculate
```

```python
x_poly=x_poly.fit_transform(x)
x_poly
```

```python
poly_regressor=LinearRegression()
```

```python
poly_regressor.fit(x_poly,y)
```

```python
new_pred=poly_regressor.predict(x_poly)
new_pred
```

```python
plt.scatter(x,y,label ="Actual Data")
plt.plot(x,new_pred , label = "Predicted Data", color = "red")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.title(" Polynomial Linear Regression")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
full_health_data = pd.read_csv("/content/drive/MyDrive/data.csv")
full_health_data = full_health_data.dropna()
x = pd.to_numeric(full_health_data["Average_Pulse"])
y = pd.to_numeric(full_health_data["Calorie_Burnage"])

slope, intercept, r,p,std_err = stats.linregress(x,y)
#slope = Slope (m) of the regressor line
#intercept = intercept(c), i.e., value of y when x=0
#r = Correlation coefficient (strength of relationship)
#p = P-value(checks if the relationship is statistically significant)
#std_err= Standard error(accuracy of the calculated slope)

plt.scatter(x,y)
plt.plot(x,slope*x+intercept)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import stats

x=[23,45,4,56,6,7,7,7,66,78]
y=[23,45,34,5,32,33,23,20,25,30] # Changed y to have the same length as x

slope, intercept, r,p,std_err = stats.linregress(x,y)

def myfunc(x):
  return slope*x+intercept

mymodel=list(map(myfunc,x))

plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show()
```
