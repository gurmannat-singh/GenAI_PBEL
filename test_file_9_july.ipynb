```python
#import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
#load dataset
df=pd.read_csv("/content/drive/MyDrive/dataset1.csv")
df
```

```python
print("Dataset Preview:")
print(df.head())
```

```python
#Check column names
print("\nColumns:", df.columns)
```

```python
#rename columns if needed
#assume first column =label, second column = message
df = df.iloc[:,:2]
df.columns=["message","label"] # Corrected: 'message' is the first column, 'label' is the second
```

```python
#convert labels to numeric
#ham = 0, spam = 1
df["label"]=df["label"].map({"ham":0,"spam":1})
```

```python
#features and target
X=df["message"]
y=df["label"]
```

```python
# Convert text into numerical vectors using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')

x_vectorized = vectorizer.fit_transform(X)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    x_vectorized,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
# Create KNN model
knn = KNeighborsClassifier(n_neighbors=5)

# Train model
knn.fit(X_train, y_train)

# Predict on test data
y_pred = knn.predict(X_test)
```

```python
# Model Accuracy
print("\nAccuracy:", accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# User Input Prediction
user_message = input("\nEnter a message: ")

# Transform input using same TF-IDF vectorizer
user_vector = vectorizer.transform([user_message])

# Predict
prediction = knn.predict(user_vector)

if prediction[0] == 1:
    print("\nPrediction: SPAM")
else:
    print("\nPrediction: HAM")
```
